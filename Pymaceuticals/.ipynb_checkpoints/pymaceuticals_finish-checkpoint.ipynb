## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_combine = pd.merge(study_results, mouse_metadata, how="left", on="Mouse ID")
# Display the data table for preview
data_combine
```

```python
# Checking the number of mice.
len(data_combine["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
data_combine.loc[data_combine.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
data_combine.loc[data_combine["Mouse ID"]=="g989"]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
data_clean = data_combine[data_combine["Mouse ID"]!="g989"]
data_clean
```

```python
# Checking the number of mice in the clean DataFrame.
len(data_clean["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.
means = data_clean.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
medians = data_clean.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
variances = data_clean.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
sds = data_clean.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sems = data_clean.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
summary_statistics = pd.DataFrame({"Mean Tumor Volume":means,
                              "Median Tumor Volume":medians,
                              "Tumor Volume Variance":variances,
                              "Tumor Volume Std. Dev.":sds,
                              "Tumor Volume Std. Err.":sems})
summary_statistics
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
summary_statistics = data_clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
final_tumor_vol = data_clean.groupby(["Mouse ID"])['Timepoint'].max()
# Start by getting the last (greatest) timepoint for each mouse
final_tumor_vol = final_tumor_vol.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
data_merge = pd.merge(final_tumor_vol, data_clean,how="left", on=["Mouse ID","Timepoint"])
data_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)

tumor_vol_treatments = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol_treatment = data_merge.loc[data_merge["Drug Regimen"]==treatment ,'Tumor Volume (mm3)']    
    
    # add subset 
    tumor_vol_treatments.append(tumor_vol_treatment)    
    
    # Determine outliers using upper and lower bounds
    quartiles = tumor_vol_treatment.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_vol_treatment.loc[(tumor_vol_treatment < lower_bound) | (tumor_vol_treatment > upper_bound)]
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol_treatments, labels = treatments)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
df4= data_clean.replace(np.nan,0)
df4 = data_clean.loc[data_clean["Drug Regimen"]=="Capomulin"]
df4 = df4.loc[df4["Mouse ID"] == "j246"]
df4.plot.line(x="Timepoint", y="Tumor Volume (mm3)")
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse j246')
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
df5 = data_clean.loc[data_clean["Drug Regimen"] == "Capomulin"]
# df5.loc[df5["Mouse ID"] == "j246"]
df5 = df5.groupby("Mouse ID").mean()
df5.plot.scatter(x='Weight (g)',
                      y='Tumor Volume (mm3)'
                     )
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Weight vs Average Tumor Volume (mm3) for Capomulin ')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr=round(st.pearsonr(df5['Weight (g)'],df5['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and the average tumor volume(mm3) is {corr}")
x_values = df5['Weight (g)']
y_values = df5['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
